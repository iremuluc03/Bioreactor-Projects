```python
# Bonus 

#  imports 
import numpy as np
from scipy.optimize import minimize_scalar

# parameters
mu_m = 0.7      # h^-1
Ks = 20.0       # mg/L
S0 = 1000.0     # mg/L
Yxs = 0.5       # gdw/gS 
alpha = 0.1
beta = 0.02
```

```python
# at steady state

# substrate concentration function
def S_ss(D):
    return Ks * D / (mu_m - D)

# biomass concentration function
def X_ss(D):
    return Yxs * (S0 - S_ss(D))

# biomass productivity XD
def biomass_productivity(D):
    return D * X_ss(D)

# product productivity PD
def product_productivity(D):
    return (alpha * D + beta) * X_ss(D)

# Optimize for 0 < D < mu_m
bounds = (1e-9, mu_m - 1e-9)

res_X = minimize_scalar(lambda D: -biomass_productivity(D), bounds=bounds, method="bounded")
res_P = minimize_scalar(lambda D: -product_productivity(D), bounds=bounds, method="bounded")

D_optimal_X = res_X.x
D_optimal_P = res_P.x

# Results
print("Biomass productivity optimum")
print(f"D_opt_X = {D_optimal_X:.3f} h^-1")
print(f"S = {S_ss(D_optimal_X):.2f} mg/L")
print(f"X = {X_ss(D_optimal_X):.2f} mg/L")
print(f"XD = {biomass_productivity(D_optimal_X):.2f} mg/L/h")

print("\nProduct productivity optimum")
print(f"D_opt_P = {D_optimal_P:.3f} h^-1")
print(f"S = {S_ss(D_optimal_P):.2f} mg/L")
print(f"X = {X_ss(D_optimal_P):.2f} mg/L")
print(f"PD = {product_productivity(D_optimal_P):.2f} mg/L/h")
```
